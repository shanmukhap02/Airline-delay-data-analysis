```python
df['CarrierDelay'] = df['CarrierDelay'].fillna(0)
df['WeatherDelay'] = df['WeatherDelay'].fillna(0)
df['NASDelay'] = df['NASDelay'].fillna(0)
df['SecurityDelay'] = df['SecurityDelay'].fillna(0)
df['LateAircraftDelay'] = df['LateAircraftDelay'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=0,how='any',inplace=True)
```

```python
df.isnull().sum()
```

```python
#date of flight
df.rename(columns={'DayofMonth':'Day'},inplace=True)
#df['Date'] = pd.to_datetime(df[['Year','Month','Day']])

df['Day'] = df['Day'].astype('str')
df['Month'] = df['Month'].astype('str')
df['Year'] = df['Year'].astype('str')
df['Date'] = df['Year'] +'/'+ df['Month'] +'/'+ df['Day']
```

```python
df = df.drop(df.columns[[0,1,2]],axis=1)
df
```

```python
datecol = df.pop('Date')

df.insert(0,'Date',datecol)
df
```

```python
df['DepTime'] = df.DepTime.astype(int)
df['ArrTime'] = df.ArrTime.astype(int)
df
```

```python
for col in ['DepTime','CRSDepTime','ArrTime','CRSArrTime' ]:
    df[col] = df[col].fillna('0000')
    df[col] = df[col].apply(lambda x: f'{int(x):04}')
    df[col] = df[col].apply(lambda x: x[:2]+':'+x[2:])
```

```python
df
```

```python
#Date time for dep,arr,crsarr,crsdep
df['DepTime'] = df['DepTime'].replace('24:00','00:00')
df['DepTime_Timestamp'] = df['Date']+' '+df['DepTime']
df['DepTime_Timestamp'] = pd.to_datetime(df['DepTime_Timestamp'])

df['ArrTime'] = df['ArrTime'].replace('24:00','00:00')
df['ArrTime_Timestamp'] = df['Date']+' '+df['ArrTime']
df['ArrTime_Timestamp'] = pd.to_datetime(df['ArrTime_Timestamp'])

df['CRSDepTime'] = df['CRSDepTime'].replace('24:00','00:00')
df['CRSDepTime_Timestamp'] = df['Date']+' '+df['CRSDepTime']
df['CRSDepTime_Timestamp'] = pd.to_datetime(df['CRSDepTime_Timestamp'])

df['CRSArrTime'] = df['CRSArrTime'].replace('24:00','00:00')
df['CRSArrTime_Timestamp'] = df['Date']+' '+df['CRSArrTime']
df['CRSArrTime_Timestamp'] = pd.to_datetime(df['CRSArrTime_Timestamp'])
```
